### generate pile uniform random weights

```python
from decimal import Decimal, getcontext

getcontext().prec = 30
total = Decimal("1")
parts = all_subdomain_num

each_part = total / Decimal(parts)

all_parts = []

for _ in range(parts - 1):
    rounded_part = each_part.quantize(Decimal('1.0000000000000000000000000000'))
    all_parts.append(rounded_part)
last_part = total - sum(all_parts)
all_parts.append(last_part)
print(all_parts)

domains = ["ArXiv", "BookCorpus2", "Books3", "DM_Mathematics", "Enron_Emails", "EuroParl", "FreeLaw", "Github", "Gutenberg_(PG-19)", "HackerNews", "NIH_ExPorter", "OpenSubtitles", "OpenWebText2", "PhilPapers", "Pile-CC", "PubMed_Abstracts", "PubMed_Central", "StackExchange", "USPTO_Backgrounds", "Ubuntu_IRC", "Wikipedia_(en)", "YoutubeSubtitles"]
subdomain_names = []
for domain in domains:
    for i in range(subdomain_num):
        subdomain_name = domain + "_" + str(i)
        subdomain_names.append(subdomain_name)

subdomain_weights = {}
for j in range(all_subdomain_num):
    subdomain_weights[subdomain_names[j]] = float(all_parts[j])
print(subdomain_weights)
```

```python
eval_pile_weights = pile_uniform["eval_domain_weights"]
print(eval_pile_weights)
```

```python
pile_random_weights = {}
pile_random_weights["train_domain_weights"] = subdomain_weights
pile_random_weights["eval_domain_weights"] = eval_pile_weights
print(pile_random_weights)
```

```python
subdomain_fw = open(pile_random_path, "w", encoding="utf-8")
json.dump(pile_random_weights, subdomain_fw)
print("=== finish writing ===")
```

### generate weights randomly

```python
import random

def generate_random_parts(num_parts):
    # 生成随机数
    random_numbers = [random.random() for _ in range(num_parts)]
    
    # 计算总和
    total = sum(random_numbers)
    
    # 归一化，使总和为1
    normalized_numbers = [num / total for num in random_numbers]
    
    return normalized_numbers

# 设置部分数量
num_parts = 220

# 生成随机部分
parts = generate_random_parts(num_parts)
print(parts)
```

```python
random_subdomain_weights = {}
for j in range(all_subdomain_num):
    random_subdomain_weights[subdomain_names[j]] = float(parts[j])
print(random_subdomain_weights)
```

```python
randomly_pile_random_weights = {}
randomly_pile_random_weights["train_domain_weights"] = subdomain_weights
randomly_pile_random_weights["eval_domain_weights"] = eval_pile_weights
print(randomly_pile_random_weights)
randomly_pile_random_path = "/home/wth/My_codes/doremi/configs/randomly_pile_uniform.json"
subdomain_fw = open(randomly_pile_random_path, "w", encoding="utf-8")
json.dump(randomly_pile_random_weights, subdomain_fw)
print("=== finish writing ===")
```
